```python
far = pd.DataFrame({
    'real': [np.array(Image.open(f'miau/{i}.png'))[240,427,1] for i in range(88)],
    'pred': [(-22/1600)*(i**2) + 81/20*i + 1 for i in range(88)]
    })
far.plot()
```

```python
def to_depth_far(color):
    a = -22/1600
    b = 81/20
    c = 1 - color
    return (-b+math.sqrt(b**2 - 4*a*c))/(2*a)
```
